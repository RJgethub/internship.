```python
# import all required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r"C:\Users\Lenovo\Downloads\Compressed\chromedriver_win32\chromedriver.exe")
```

```python
url = 'https://bit.ly/3HbOtY7'
driver.get(url)
```

```python
driver.find_element_by_xpath("//div[@class='float-right TBxAN']").click()
```

```python
#Creating empty lists
brand=[]
year=[]
fuel=[]
transmission=[]
km=[]
price=[]
prod_URL=[]
```

```python
#Getting the URL of the cars
for i in range(0,500):
    url=driver.find_elements_by_xpath("//div[@class='col-4']/a")
    for i in url:
        prod_URL.append(i.get_attribute('href'))
    try:
        next_btn = driver.find_element_by_xpath("//button[@class='btn _3BmFU AgO6q']").click()
    except NoSuchElementException:
        pass
```

```python
prod_URL
```

```python
for i in prod_URL:
    driver.get(i) 
    
#Extracting car name
    try:
        brand.append(driver.find_element_by_xpath("//h2[@class='_3FpCg']").text)
    except NoSuchElementException as e:
        brand.append("-")
        
#Extracting the transmission
    try:
        transmission.append(driver.find_element_by_xpath("//p[@class='cvakB']/span").text)  
    except NoSuchElementException as e:
        transmission.append("-")
        
#Extracting the fuel consumed
    try:
        fuel.append(driver.find_element_by_xpath("//ul[@class='bVR0c']/li[3]").text)
    except NoSuchElementException as e:
        fuel.append("-")
        
#Extracting km driven
    try:
        km.append(driver.find_element_by_xpath("//ul[@class='bVR0c']/li[1]").text)
    except NoSuchElementException as e:
        km.append("-") 
        

        #Extracting the price details
    try:
        price.append(driver.find_element_by_xpath("//div[@class='_7udZZ']").text.replace('',''))
    except NoSuchElementException as e:
        price.append("-")
```

```python
#Creating a dataframe and checking the data extracted
cars24 = pd.DataFrame({'name':brand,'selling_price':price,'km_driven':km,'fuel':fuel,'transmission':transmission})
```
